# Assignment 3 - Matplotlib

```python
import matplotlib.pyplot as plt
```

## Use any 2 types of charts to visualize some data

#### you can use datasets or you can write your own data

```python
labels = ['A', 'B', 'C', 'D']
values = [20, 35, 42, 13]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(labels, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
